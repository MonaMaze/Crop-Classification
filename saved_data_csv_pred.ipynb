```python
# Load all libraries
import rasterio
import numpy as np
import pandas as pd
import os
from collections import Counter
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Create the column names of dataframe from indices' names
colNames = []
folder = 'CC'
i = 0
for subfolder in os.listdir(folder):
    for filename in os.listdir(os.path.join(folder, subfolder)):
        img_name = filename.split('_')[0]
        colNames.append(img_name)
    i += 1
    if i > 0:
        break
```

```python
colNames
```

```python
df = pd.DataFrame(columns = colNames)
```

```python
df
```

```python
# Get the values of images in one dataframe
subfolder = '2023-07-27'
x = len([name for name in os.listdir(os.path.join(folder, subfolder)) if os.path.isfile(os.path.join(folder, subfolder, name))])
i = 0
array = {}
for filename in os.listdir(os.path.join(folder, subfolder)):
    if i < x:
        img = rasterio.open(os.path.join(folder, subfolder, filename))
        array[f'var{i}'] = img.read()
        i += 1
arra = np.concatenate(list(array.values()),0)
df_img = pd.DataFrame(arra.reshape([x,-1]).T, columns=colNames)
df = pd.concat([df, df_img])
```

```python
import gc
del img
del array
del arra
del df_img
gc.collect()
```
